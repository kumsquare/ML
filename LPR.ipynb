```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
data=pd.read_csv('Position_Salaries.csv')
x=data.iloc[:,1:2].values
y=data.iloc[:,2].values
print(x,y)
linreg=LinearRegression()
linreg=linreg.fit(x,y)

plt.scatter(x,y,color='blue')
plt.plot(x,linreg.predict(x),color='red')
plt.xlabel('Position level')
plt.ylabel('Salaries')
plt.title('Bluff detection model(Linear Regression)')
plt.show()

polyreg=PolynomialFeatures(degree=3)
x_poly=polyreg.fit_transform(x)
linreg2=LinearRegression()
linreg2=linreg2.fit(x_poly,y)

plt.scatter(x,y,color='blue')
plt.plot(x,linreg2.predict(x_poly),color='red')
plt.xlabel('Position level')
plt.ylabel('Salaries')
plt.title('Bluff detection model(Polynomial Regression)')
plt.show()

lin_pred=linreg.predict([[11]])
poly_pred=linreg2.predict(polyreg.fit_transform([[11]]))
print(lin_pred)
print(poly_pred)
```
