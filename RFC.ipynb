```python
from sklearn import datasets 
import pandas as pd 
# Loading the iris plants dataset (classification) 
iris= datasets.load_iris() 
from sklearn.ensemble import RandomForestClassifier 
# creating a RF classifier 
clf = RandomForestClassifier(n_estimators = 100) 
# creating dataframe of IRIS dataset 
data = pd.DataFrame({'sepallength': iris.data[:, 0], 'sepalwidth': iris.data[:, 1],'petallength': 
iris.data[:, 2], 'petalwidth': iris.data[:, 3],'species': iris.target}) 
print(data) 
print('Taget names:',iris.target_names) 
# Training the model on the training dataset 
# fit function is used to train the model using the training sets as parameters 
# Spliting arrays or matrices into random train and test subsets 
from sklearn.model_selection import train_test_split 
x,y= datasets.load_iris(return_X_y=True) 
print(x,y) 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30) 
y_pred=clf.fit(x_train, y_train).predict(x_test) 
from sklearn import metrics 
print('ACCURACY OF THE MODEL: ', metrics.accuracy_score(y_test, y_pred)*100) 
# predicting the type of flower for given test data: 
guess=clf.predict([[4, 5, 1, 2]]) 
print('Prediction for the test data is',iris.target_names[guess]) 
```
